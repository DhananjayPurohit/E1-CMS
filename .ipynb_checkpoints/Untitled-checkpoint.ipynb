```python
f = open('out_text1.txt', 'r')
l = []
t = False
for i in f.readlines():
    if "Unit" in i:
        t = True
        continue
    if t:
        if "Unit" in i:
            break
        i = i.replace("\n","")
        i = i.replace("  ","")
        l.append(i)

k = []
for i in l:
    if i == "" or i == " ":
        continue
    k.append(i)
k
```
